Funciones de bordes

Generador de bordes 1 pixel

```python
def genera_bordes (imagen_seg):

    """Función que a partir de una imagen segmentada devuelve una
    imagen que solo contiene las lineas frontera entre los segmentos. Debemos tener importada la librería
    numpy y pandas """
    
    import numpy as np
    import pandas as pd
    
    (alto,ancho)=imagen_seg.shape #tomamos el tamaño de la imagen de entrada
    imagen_lin=np.zeros((alto,ancho))
    
    
    for f in range(alto):
        for c in range(ancho):
            if f==0:
                if c==0:
                    pass
                else:
                    if imagen_seg[f][c]!=imagen_seg[f][c-1] : #si el pixel es diferente al anteiror es borde
                        imagen_lin[f][c-1]=1.0 #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    else: pass

            elif c==0:
                if imagen_seg[f][c]!=imagen_seg[f-1][c] :  #si el pixel es diferente a su superior es borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                else: pass
            else:
                if imagen_seg[f][c]!=imagen_seg[f][c-1] and imagen_seg[f][c]!=imagen_seg[f-1][c] and imagen_seg[f][c]!=imagen_seg[f-1][c-1]: #si el punto es diferente al anterior o al superior es borde
                    imagen_lin[f][c-1]=1.0  #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c-1]=1.0
                elif imagen_seg[f][c]!=imagen_seg[f][c-1] and imagen_seg[f][c]!=imagen_seg[f-1][c] : #si el punto es diferente al anterior o al superior es borde
                    imagen_lin[f][c-1]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                elif imagen_seg[f][c]!=imagen_seg[f][c-1]:
                    imagen_lin[f][c-1]=1.0
                elif imagen_seg[f][c]!=imagen_seg[f-1][c]:
                    imagen_lin[f-1][c]=1.0
    return imagen_lin
        
```

```python
def genera_bordes2 (imagen_seg):

    """Función que a partir de una imagen segmentada devuelve una
    imagen que solo contiene las lineas frontera entre los segmentos de 2 pixeles de ancho. Debemos tener importada la librería
    numpy y pandas """

    import numpy as np
    import pandas as pd
    
    (alto,ancho)=imagen_seg.shape #tomamos el tamaño de la imagen de entrada
    imagen_lin=np.zeros((alto,ancho))
    
    
    
    for f in range(alto):
        for c in range(ancho):
            if f==0:
                if c==0:
                    pass
                else:
                    if imagen_seg[f][c]!=imagen_seg[f][c-1] : #si el pixel es diferente al anteiror es borde
                        imagen_lin[f][c-1]=1.0 #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                        imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
                    else: pass
                    
            elif c==0:
                if imagen_seg[f][c]!=imagen_seg[f-1][c] :  #si el pixel es diferente a su superior es borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
                else: pass
            else:
                if imagen_seg[f][c]!=imagen_seg[f][c-1] and imagen_seg[f][c]!=imagen_seg[f-1][c] and imagen_seg[f][c]!=imagen_seg[f-1][c-1]: #si el punto es diferente al anterior o al superior es borde
                    imagen_lin[f][c-1]=1.0  #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c-1]=1.0
                    imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
                elif imagen_seg[f][c]!=imagen_seg[f][c-1] and imagen_seg[f][c]!=imagen_seg[f-1][c] : #si el punto es diferente al anterior o al superior es borde
                    imagen_lin[f][c-1]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f-1][c]=1.0   #marcamos en la imagen_lin el ultimo punto con mismo valor como borde
                    imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
                elif imagen_seg[f][c]!=imagen_seg[f][c-1]:
                    imagen_lin[f][c-1]=1.0
                    imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
                elif imagen_seg[f][c]!=imagen_seg[f-1][c]:
                    imagen_lin[f-1][c]=1.0
                    imagen_lin[f][c]=1.0 #marcamos el punto actual tambien como borde
    return imagen_lin
```

Funcion que calcula los bordes como combinación de hasta 5 imagenes de bordes.

```python
def calcula_bordes_5 (treshold, imagen1=[],imagen2=[],imagen3=[],imagen4=[],imagen5=[]):
    
    """Función que dadas 5 imagenes con la misma dimensión, unidimensionales de bordes siendo 0 el valor no borde y 1 el valor de borde,
    devuelve una predicción del borde real teniendo en cuenta todas las segmentaciones y el nivel entre 0 y 100 requerido"""
    
    import numpy as np
    import pandas as pd
    
    #poner control de fotos no iguales
    
    datos=[]
    bordes=[]
    imagenes_validas=5
    if len(imagen1)==0:
        print("Datos no válidos")
        return
    if len(imagen2)==0:
        imagen2=np.zeros((imagen1.shape))
        imagenes_validas=imagenes_validas - 1
    if len(imagen3)==0:
        imagen3=np.zeros((imagen1.shape))
        imagenes_validas=imagenes_validas - 1
    if len(imagen4)==0:
        imagen4=np.zeros((imagen1.shape))
        imagenes_validas=imagenes_validas - 1
    if len(imagen5)==0:
        imagen5=np.zeros((imagen1.shape))
        imagenes_validas=imagenes_validas - 1
        
    #comprobamos que las imagenes son iguales
    if imagen1.shape==imagen2.shape==imagen3.shape==imagen4.shape==imagen5.shape:pass
    else:
        print ("las imagenes no son iguales")
        return
    bordes=imagen1+imagen2+imagen3+imagen4+imagen5
    bordes=bordes/imagenes_validas
    interseccion=np.where(bordes < treshold,0,1)
    
    
    return bordes,interseccion
    
    
```
