```python
def carga_fotos (imagen_rgb,imagen_mat,directorio=""):
    """funci√≥n que carga fotos RGB y provenientes de Matlab """
    
    from skimage import io
    from scipy.io import loadmat
    import os
    
    # Cargamos la imagen de directorio base
    
    image = io.imread(os.path.join(directorio,imagen_rgb))

    # Cargamos datos de Matlab de directorio base
    mat_data = loadmat(os.path.join(directorio,imagen_mat))
    edges = mat_data['groundTruth'][0][0][0][0][1] #bordes
    segmap = mat_data['groundTruth'][0][0][0][0][0] #imagen segmentada
    
#     # Cargamos la imagen de directorio test
#     test="imagenes_test"
#     image = io.imread(os.path.join(test,imagen_rgb))

#     # Cargamos datos de Matlab de directorio test
#     mat_data = loadmat(os.path.join(test,imagen_mat))
#     edges = mat_data['groundTruth'][0][0][0][0][1] #bordes
#     segmap = mat_data['groundTruth'][0][0][0][0][0] #imagen segmentada   
    

#     plt.figure(figsize=(20, 10))
#     plt.subplot(3, 1, 1)
#     plt.title('Imagen Original')
#     plt.imshow(image)
#     plt.axis('off')

#     plt.figure(figsize=(20, 10))
#     plt.subplot(3, 1, 2)
#     plt.title('datos1')
#     plt.imshow(edges,cmap="grey")
#     plt.axis('off')

#     plt.subplot(3, 1, 3)
#     plt.title('datos2')
#     plt.imshow(segmap)
#     plt.axis('off')
    
    return image,segmap,edges

```
